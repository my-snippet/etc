이산시간 푸리에 변환(DFTP)
===
이산 푸리에 변환(discrete Fourier transform, DFT)은 이산적인 입력 신호에 대한 푸리에 변환으로, 디지털 신호 분석과 같은 분야에 사용된다.
이산 푸리에 변환은 고속 푸리에 변환을 이용해 빠르게 계산할 수 있다.
- 위키 백과 : https://ko.wikipedia.org/wiki/%EC%9D%B4%EC%82%B0_%ED%91%B8%EB%A6%AC%EC%97%90_%EB%B3%80%ED%99%98-


모든 n에 대한 $x(n)$의 합이 무한이 아니면(absolutely summable) 이산 푸리에 변환을 아래와 같이 정의한다.

$X(e^{jw}) = F[x(n)] = \sum^{\infty}_{n=-\infty} x(n)e^{-jwn}$ 
---

한편, 이것을 극좌표 방식으로 나타내면

$X(e^{jw}) = |X(e^{jw})|e^{j\angle X(e^{jw})}$
---

두 가지 중요한 성질
===
1. 주기성
---
이산시간 푸리에 변환$X(e^{jw})$은 주기 2$\pi$로 $\omega$에 대해 주기적이다.

$X(e^{jw}) = X(e^{j(w + 2\pi)})$
---
2. 대칭성
---
실수 $x(n)$에 대해 $X(e^{jw})$은 켤레 대칭이다.
$X(e^{-jw}) = X^*(e^{jw})$
---


이산시간 푸리에 변환(DFTP)의 성질
===
1.linearity : 함수에 인자를 통째로 넣은 것과 여러 함수로 분리해서 합한 값은 같다.

2.shift : 시간 이동은 곧 위상 이동이다.

3.주파수 이동 : 복소지수의 곱이 곧 주파수 이동이다.

4.conjugation : 시간 영역에서의 켤레는 주파수 영역에서의 반전과 켤레에 해당한다.

5.folding : 시간 영역에서의 반전은 주파수에서도 반전이다.

6.symmetry property : 모든 실수 함수는 우함수와 기함수의 합으로 분리된다.

7.convolution : 시간 영역에서 주파수 응답을 알기위한 합성곱이다.

8.곱셈 : 시간에서의 곱은 주파수의 convolution이다 $\rightarrow$ duality

9.주파수 미분 : $F[nx(n)] = j\frac{dX(e^{jw})}{dw}$

10.parseval's theorem : 신호 x(n)의 에너지로부터 에너지 밀도 스펙트럼이 정의된다.

LTI 시스템의 주파수 표현
===
1.복소지수 신호 : $x(n) = e^{jw_0n} \rightarrow H(e^{jw}) \rightarrow y(n) = H(e^{jw_0n})$

2.임의의 신호 : $x(e^{jw}) \rightarrow H(e^{jw}) \rightarrow y(e^{jw}) = H(e^{jw})X(e^{jw})$

이산 신호 샘플링 및 복원
===
샘플링
---
연속시간의 주파수 변환으로 부터 샘플링 간격을 계산한다.

$\omega = \Omega T_s$
---

복원
---
샘플을 impulse sequence(train)으로 전환한다. 이것은 기존의 $\sum$의 형태를 풀어쓴 것이다. 그리고 그것들을 LPF에 통과시키면 완성된다.

과정
---
Zero-order-hold -> First-order-hold -> Spline 보간
